# Preprocessing of airports_terms_of_lease.csv

```python
import numpy as np
import pandas as pd

import csv
```

```python
# load data
airports_terms_of_lease = pd.read_csv('../data/raw/airports_terms_of_lease.csv', quoting=csv.QUOTE_NONE, dtype='str')
airports_terms_of_lease
```

```python
# replace quotes in header and data
airports_terms_of_lease = airports_terms_of_lease.applymap(lambda x: x.replace('"', ''))
airports_terms_of_lease.columns = airports_terms_of_lease.columns.str.replace('"', '')

# cast dtypes
airports_terms_of_lease[['sqm of store', 'price per sqm/month']] = airports_terms_of_lease[['sqm of store', 'price per sqm/month']].astype('int16')

print(airports_terms_of_lease.dtypes)

airports_terms_of_lease
```

```python
# save processed data
airports_terms_of_lease.to_csv('../data/processed/airports_terms_of_lease.csv')
```
